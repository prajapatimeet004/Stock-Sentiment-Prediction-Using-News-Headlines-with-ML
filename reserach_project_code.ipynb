```python
import pandas as pd
data = pd.read_csv(r"C:\Users\Admin\OneDrive\Desktop\research\data_final.csv")
data.head()
```

```python
#import spacy for nlp task
import spacy as sp
nlp = sp.load('en_core_web_sm')
```

```python
#cleaning the data by making the function that lemmatizie the data,remove stop words,alpha and url
def clean_text(text):
    doc = nlp(text)
    dat = [tokens.lemma_.lower() for tokens in doc if not tokens.is_stop and not tokens.is_punct and not tokens.like_url]
    
    return ' '.join(dat)

data['text'] = data['text'].apply(clean_text)
```

```python
#spliting data into training and testing 

x = data['text']
y = data['label']

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.30)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()
x_train_vec = vectorizer.fit_transform(x_train)# Transform text to numerical
x_test = vectorizer.transform(x_test)
```

```python
#importing models to test which one is best for this
from sklearn.tree import DecisionTreeClassifier
DTC = DecisionTreeClassifier(criterion='gini',max_depth=3,splitter='best')
DTC.fit(x_train_vec,y_train)
```

```python
#classifing the value
y_pred = DTC.predict(x_test)
print(y_pred)
```

```python
#classification report 
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred)) #getting accurecy of 68% because class is unbalanced
```

```python
#using knn classifier

from sklearn.neighbors import KNeighborsClassifier
KNN = KNeighborsClassifier(n_neighbors=3)
KNN.fit(x_train_vec,y_train)
#classifing the value
y_pred = KNN.predict(x_test)
print(y_pred)
#classification report 
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
#using svm classifier

from sklearn.svm import SVC
svm = SVC(kernel='poly')
svm.fit(x_train_vec,y_train)
#classifing the value
y_pred = svm.predict(x_test)
print(y_pred)
#classification report 
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))

#svm is performing best amoung all algo
```
